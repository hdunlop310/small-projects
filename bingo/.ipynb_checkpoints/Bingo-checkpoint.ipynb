```python
#bingo 
import random, numpy
def generate_card():
    number = 25
    max_number = 50
    array = []
    while len(array) < number:
        new_num = random.randint(0,max_number)
        if new_num not in array:
            array.append(new_num)

    arr = numpy.array(array)

    bingo_card = arr.reshape(5, 5)
    return bingo_card

def number_caller():
    game = True
    number_called = []
    max_number = 50
    while game and len(number_called) < 50: 
        num = random.randint(0,max_number)
        if num not in number_called:
            print(num)
            number_called.append(num)
        if len(number_called) == 50:
            game = False
        bingo = input("Did a player reach bingo? Enter 'Y' or 'N' ")
        while bingo.upper != 'Y' or bingo.upper != 'N':
            bingo = input("Invalid input. Did a player reach bingo? Enter 'Y' or 'N' ")
        if bingo.upper() == 'Y':
            game = False
    
        
num_of_players = int(input("How many players? "))

for counter in range(0,num_of_players):
    print(generate_card())
    print()
    print()

print("Time to start the game!")
number_caller()
```
